## Bonus Tasks
Congratulations! You have created a csv file, which we can use to explore our extracted data. So far, the file, however, only features the most recent articles, i.e. the articles from page 1 of the Sport's Features section. Write a program that creates a csv file with tag, title, and year of all articles from 2021-2019. Hint: It might be easier to first add some articles from 2018 and then remove them from the file.


```python
article_previews=[]
for x in range(1, 10):
    page = requests.get("https://www.thecrimson.com/tag/sports-features/page/{0}/".format(x))
    soup = BeautifulSoup(page.text, "html.parser")
    article_list = soup.find_all("div", class_="css-7c03k-CleanParagraphs")
    for element in article_list:
        article_previews.append(element)
    
print(article_previews)
```

```python
file_write = open('article_previews.csv', 'w', newline='')

writer = csv.writer(file_write)

for preview in article_previews:
    date_string_temp = re.findall("[\d]{4}", str(preview.find('a').get('href')))
    preview_line = [preview.find("div", class_="css-14eqys5-So").text, preview.find("h1").text, date_string_temp[0]]
    writer.writerow(preview_line)

# close the files

file_write.close()
```
